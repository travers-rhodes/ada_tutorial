```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

file_names_labels = [("../final_results/rice_constant_camera.txt", "Rice"),
                     ("../final_results/rice_constant_camera2.txt", "Rice"),
                     #("../final_results/rice_constant_camera3.txt", "Rice"), need to redo rice dried
                     ("../final_results/rice_constant_camera3_redo.txt", "Rice"), #needed to redo rice dried
                    ("../final_results/rice_constant_no_camera.txt", "Rice No Camera"),
                    #("../final_results/rice_constant_no_camera2_redo.txt", "Rice No Camera"), needed to redo because rice dried
                    ("../final_results/rice_constant_no_camera2_redo2.txt", "Rice No Camera"), #needed to redo because rice dried
                    ("../final_results/rice_constant_no_camera3.txt", "Rice No Camera"),
                     ("../final_results/rice_random_with_camera.txt", "Rice Random"),
                    ("../final_results/rice_random_with_camera2.txt", "Rice Random"),
                    ("../final_results/rice_random_with_camera3.txt", "Rice Random"),
                    ("../final_results/rice_random_no_camera.txt", "Rice Random No Camera"),
                    ("../final_results/rice_random_no_camera2.txt", "Rice Random No Camera"),
                    #("../final_results/rice_random_no_camera3.txt", "Rice Random No Camera"), needed to redo because rice had dried out overnight
                    ("../final_results/rice_random_no_camera3_redo.txt", "Rice Random No Camera"),
                    ]

scoop_dist = 1
# one direction to/from mouth
mouth_dist = 10

plt.figure(figsize=(10,8))
for nameAndLabel in file_names_labels:
  dat = np.loadtxt(nameAndLabel[0], delimiter=",", usecols = (0,1,2,3))
  plt.plot(np.arange(dat.shape[0]) * 2 * mouth_dist + mouth_dist + dat[:,3] * scoop_dist, dat[:,1], label=nameAndLabel[1])
    
plt.legend(loc='upper left')
plt.xlim((0,500))
plt.ylabel("Cumulative Mass Fed (g)")
plt.xlabel("Cumulative Distance Traveled (m)")

plt.show()
#plt.savefig("ValueOfCameraFeedback")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

file_names_labels = [("../final_results/nuts_constant_camera.txt", "Rice"),
                     ("../final_results/nuts_constant_camera2.txt", "Rice"),
                     ("../final_results/nuts_constant_camera3.txt", "Rice"),
                    ("../final_results/nuts_constant_no_camera.txt", "Rice No Camera"),
                    ("../final_results/nuts_constant_no_camera2.txt", "Rice No Camera"),
                    ("../final_results/nuts_constant_no_camera3.txt", "Rice No Camera"),
                    ("../final_results/nuts_random_with_camera.txt", "Rice Random"),
                    ("../final_results/nuts_random_with_camera2.txt", "Rice Random"),
                    ("../final_results/nuts_random_with_camera3.txt", "Rice Random"),
                    ("../final_results/nuts_random_no_camera.txt", "Rice Random No Camera"),
                    ("../final_results/nuts_random_no_camera2.txt", "Rice Random No Camera"),
                    ("../final_results/nuts_random_no_camera3.txt", "Rice Random No Camera"),
                    ]

plt.figure(figsize=(10,8))
for nameAndLabel in file_names_labels:
  dat = np.loadtxt(nameAndLabel[0], delimiter=",", usecols = (0,1,2))
  plt.plot(range(dat.shape[0]), dat[:,1], label=nameAndLabel[1])
    
plt.legend(loc='upper left')
plt.xlim((0,20))
plt.ylabel("Cumulative Mass Fed (g)")
plt.xlabel("Cumulative Distance Traveled (m)")

plt.show()
#plt.savefig("ValueOfCameraFeedback")
```

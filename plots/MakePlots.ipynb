```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

file_names_labels = [("../random_move_with_camera.txt", "Nuts with Camera Feedback"),
                     ("../random_move_no_camera.txt", "Nuts"),
                     ("../rice_random_move_with_camera.txt", "Rice with Camera Feedback"),
                       ("../rice_random_move_no_camera.txt", "Rice")]

plt.figure(figsize=(8,4))
for nameAndLabel in file_names_labels:
  dat = np.loadtxt(nameAndLabel[0], delimiter=",")
  plt.plot(dat[:,2], dat[:,1], label=nameAndLabel[1])
    
plt.legend(loc='upper left')
plt.xlim((0,60))

plt.show()
plt.savefig("ValueOfCameraFeedback")
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

file_name_templates = ["../%srandom_move_no_camera.txt",
                       "../%srandom_move_with_camera.txt",
                       "../%sconstant_move_no_camera.txt",
                       "../%sconstant_move_with_camera.txt"]

for template in file_name_templates:
  name = template % "rice_"
  dat = np.loadtxt(name, delimiter=",")
  plt.plot(dat[:,2], dat[:,1])

plt.show()
```

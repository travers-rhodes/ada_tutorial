```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'font.size': 22})


file_names_labels = [
                     ("../final_results/rice_random_with_camera.txt", "Random with Camera", "g"),
                    ("../final_results/rice_random_with_camera2.txt", None, "g"),
                    ("../final_results/rice_random_with_camera3.txt", None, "g"),
                    ("../final_results/rice_random_no_camera.txt", "Random No Camera", "r"),
                    ("../final_results/rice_random_no_camera2.txt", None, "r"),
                    #("../final_results/rice_random_no_camera3.txt", "Rice Random No Camera"), needed to redo because rice had dried out overnight
                    ("../final_results/rice_random_no_camera3_redo.txt", None, "r"),
                    ("../final_results/rice_constant_camera.txt", "Constant with Camera", "g--"),
                    ("../final_results/rice_constant_camera2.txt", None, "g--"),
                    #("../final_results/rice_constant_camera3.txt", "Rice"), need to redo rice dried
                    ("../final_results/rice_constant_camera3_redo.txt", None, "g--"), #needed to redo rice dried
                    ("../final_results/rice_constant_no_camera.txt", "Constant No Camera", "r--"),
                    #("../final_results/rice_constant_no_camera2_redo.txt", "Rice No Camera"), needed to redo because rice dried
                    ("../final_results/rice_constant_no_camera2_redo2.txt", None, "r--"), #needed to redo because rice dried
                    ("../final_results/rice_constant_no_camera3.txt", None, "r--"),
                    ]

scoop_dist = 0.32
# one direction to/from mouth
mouth_dist = 0.49

plt.figure(figsize=(12,10))
for nameAndLabel in file_names_labels:
  dat = np.loadtxt(nameAndLabel[0], delimiter=",", usecols = (0,1,2,3))
  # fill since some ended with scoops and no dumps
  dat = np.concatenate((dat,[[None, dat[-1,1], 100,100]]))
  plt.plot(np.arange(dat.shape[0]) * 2 * mouth_dist + mouth_dist + dat[:,3] * scoop_dist, 
           dat[:,1], 
           nameAndLabel[2],
           label=nameAndLabel[1])
    
plt.legend(loc='upper left')
plt.xlim((0,25))
plt.ylabel("Cumulative Mass Fed (g)")
plt.xlabel("Cumulative Distance Traveled (m)")


plt.savefig("RiceCumulativeMass",bbox_inches='tight')

plt.figure(figsize=(12,10))
format = "go"
plt.xlim((0,20.5))
plt.ylim((0,13))
camera_random_weight = []
no_camera_random_weight = []
for i,nameAndLabel in enumerate(file_names_labels):
  
  dat = np.loadtxt(nameAndLabel[0], delimiter=",", usecols = (0,1,2,3))
  scoop_masses = np.concatenate((dat[0:1,1], np.diff(dat[:,1])))
  if i < 3:
    format = "go"
    camera_random_weight.append(scoop_masses[0:20,])
  elif i < 6:
    format = "ro"
    no_camera_random_weight.append(scoop_masses[0:20,])
  elif i == 6:
    break
  #print(scoop_masses.shape)
  #print(dat.shape)
  # fill since some ended with scoops and no dumps
  #dat = np.concatenate((dat,[[None, dat[-1,1], 100,100]]))
  plt.plot(np.arange(dat.shape[0]) + 1, 
           scoop_masses, 
           format,
           label=nameAndLabel[1])

plt.xlabel("Bite Number")
plt.ylabel("Mass of Food Delivered (m)")
plt.legend(loc='upper right')
plt.legend(numpoints=1)
plt.savefig("RiceScoopSizes",bbox_inches='tight')

from scipy import stats
print("camera mean weight %s"% np.mean(np.array(camera_random_weight)))
print("no camera mean weight %s"% np.mean(np.array(no_camera_random_weight)))
stats.ttest_rel(np.array(camera_random_weight).flatten(),np.array(no_camera_random_weight).flatten())
```
